```


#hide
def colorizar(source_url, render_factor, compare=True):
    image_path = colorizer.plot_transformed_image_from_url(
                                                        url=source_url, 
                                                        render_factor=render_factor, 
                                                        compare=compare,                                                       
                                                        watermarked=False,
                                                        #figsize=(8,8)
                                                        )
    return image_path

```
